### Задание 1.2

```python
t = np.linspace(0, 10, 51)
f = np.cos(t)

title_font = {
    "fontsize": 16,
    "fontweight": "bold",
    "color": "#0000FF",
    "family": "sans",
}
label_font = {
    "fontsize": 9,
    "family": "sans-serif",
}

plt.title("График f(t)", fontdict=title_font)
plt.xlabel("Значения t", fontdict=label_font)
plt.ylabel("Значения f", fontdict=label_font)
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.plot(t, f, color="green")
plt.show()
```

```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)
plt.show()
```
