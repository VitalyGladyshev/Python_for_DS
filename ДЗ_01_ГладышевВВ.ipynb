```python
A = np.array([[3, -2, 5],
             [7, 4, -8],
             [5, -3, -4]])
B = np.array([[7],
              [3],
              [-12]])
```

```python
print(np.linalg.det(A))
```

```python
X = np.linalg.inv(A)@B
```

```python
print(X)
```

## Задание 4

```python
x = np.linspace(0, 10, 100)

y1 = np.sin(x)
y2 = np.sin(x*2)

plt.title("Графики функций")
plt.xlabel("Переменная x")

plt.plot(x, y1)
plt.plot(x, y2)

plt.legend(labels=["y1=sin(x)", "y2=sin(x*2)"], loc="lower right")
plt.show()
```
