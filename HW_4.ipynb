## Агенты

```python
def rock_agent(observation, configuration):
    """
    Только камень
    """
    return 0
```

```python
def paper_agent(observation, configuration):
    """
    Только бумага
    """
    return 1
```

```python
def scissors_agent(observation, configuration):
    """
    Только ножницы
    """
    return 2
```

```python
def random_agent(observation, configuration):
    """
    Случайный ход
    """
    return random.randrange(0, 3)
```

```python
def rock_paper_agent(observation, configuration):
    """
    Камень или бумага
    """
    return random.randint(0, 1)
```

```python
def paper_scissors_agent(observation, configuration):
    """
    Бумага или ножницы
    """
    return random.randint(1, 2)
```

```python
def rock_scissors_agent(observation, configuration):
    """
    Камень или ножницы
    """
    if random.randint(0, 1):
        return 2
    else:
        return 0
```

```python
def prev_opponent_agent(observation, configuration):
    """
    Повторяет предыдущий ход опонента
    """
    if observation.step > 0:
        return observation.lastOpponentAction
    # Первый шаг
    else:
        return random.randrange(0, 3)
```

```python
def cycle_moves_agent(observation, configuration):
    """
    Все по очереди
    """
    return observation.step % 3
```

```python
def beat_last_move_agent(observation, configuration):
    """
    Побить предыдущий ход опонента
    """
    if observation.step > 0:
        return (observation.lastOpponentAction + 1) % 3
    else:
        return random.randrange(0, 3)
```

```python
def beat_last_randomly_agent(observation, configuration):
    """
    Побить предыдущий ход опонента или случайный
    """
    if observation.step > 0:
        return (random.choice([(observation.lastOpponentAction + 1) % 3, random.randrange(0, 3)]))
    else:
        return random.randrange(0, 3)
```

```python
def ret_agent(observation, configuration):
    if observation.step > 0:
        return (observation.lastOpponentAction + 2) % 3
    else:
        return random.randrange(0, 3)
```

## Соревнование

```python
env = make("rps", configuration={"episodeSteps": 100}, debug=True)
```

```python
agents = [
    "rock_agent",
    "paper_agent",
    "scissors_agent",
    "random_agent",
    "rock_paper_agent",
    "paper_scissors_agent",
    "rock_scissors_agent",
    "prev_opponent_agent",
    "cycle_moves_agent",
    "beat_last_move_agent",
    "beat_last_randomly_agent",
    "ret_agent"
]

agents_dict = {
    "rock_agent": rock_agent,
    "paper_agent": paper_agent,
    "scissors_agent": scissors_agent,
    "random_agent": random_agent,
    "rock_paper_agent": rock_paper_agent,
    "paper_scissors_agent": paper_scissors_agent,
    "rock_scissors_agent": rock_scissors_agent,
    "prev_opponent_agent": prev_opponent_agent,
    "cycle_moves_agent": cycle_moves_agent,
    "beat_last_move_agent": beat_last_move_agent,
    "beat_last_randomly_agent": beat_last_randomly_agent,
    "ret_agent": ret_agent
}

n_agents = len(agents)
```

```python
results = {}
```

```python
for i in range(n_agents):
    for j in range(i + 1, n_agents):
        print(f"{agents[i]} против {agents[j]}")

        outcome = env.run([agents_dict[agents[i]], agents_dict[agents[j]]])
        if outcome[-1][0]['reward'] > outcome[-1][-1]['reward']:
            winner = agents[i]
        elif outcome[-1][0]['reward'] < outcome[-1][-1]['reward']:
            winner = agents[j]
        else:
            winner = 'Ничья'

        results[f'Агент {agents[i]} против Агента {agents[j]}'] = winner
```
