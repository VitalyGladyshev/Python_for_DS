```python
a = np.array([[1,6],[2,8],[3,11],[3,10],[1,7]])
print(a)
mean_a = a.mean(axis=0)
print(mean_a)
a_centered = a - mean_a
print(a_centered)
a_centered_sp = a_centered[:,0].dot(a_centered[:,1])
print(a_centered_sp)
a_centered_sp = a_centered[:,0] @ a_centered[:,1]
print(a_centered_sp)
```
